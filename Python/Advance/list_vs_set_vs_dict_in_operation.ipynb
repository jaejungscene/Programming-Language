### set is faster than list when checking existing, while list is faster than set when iterating

```python
arr = []
hashmap = {}
hashset = set()

for i in range(40000000):
    arr.append(i)
    hashmap[i] = i
    hashset.add(i)
```

```python
"""
{Fast <---------------------------------------> Slow}
set => hashmap.keys > hashmap > hashmap.values > list
"""
import time
result = {}
findNum = 40000

# start = time.time()
# if findNum in arr:
#     print("list true")
# end = time.time()
# print(f'{end-start:.10f}')
# result["list"] = end-start


start = time.time()
if findNum in hashmap:
    print("hashmap true")
end = time.time()
print(f'{end-start:.10f}')
result["hashmap"] = end-start


start = time.time()
if findNum in hashmap.keys():
    print("hashmap.keys true")
end = time.time()
print(f'{end-start:.10f}')
result["hashmap.keys"] = end-start


# start = time.time()
# if findNum in hashmap.values():
#     print("hashmap.values true")
# end = time.time()
# print(f'{end-start:.10f}')
# result["hashmap.values"] = end-start


start = time.time()
if findNum in hashset:
    print("set true")
end = time.time()
print(f'{end-start:.10f}')
result["set"] = end-start



import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(result))
years = list(result.keys())
values = list(result.values())

plt.xlabel('name')
plt.ylabel('time')
plt.bar(x, values, width=0.4)
plt.xticks(x, years, rotation=45)

plt.show()
```
