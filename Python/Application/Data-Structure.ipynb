### 문제 2: n에서 시작하는 Collatz 시퀀스의 길이를 계산합니다.

#### **참고사항:**
##### 1. Collatz sequence?

> "우박수" 또는 "우박 수열"

> 숫자가 커졌다 작아졌다를 반복하다 결국 1에 수렴하는 걸 비구름에서 빗방울이 오르락내리락하며 우박이 되는 모습에 빗대어 부르기 시작.

> 이 수열을 모든 자연수 n에 대해 유한번 재귀 반복하면 1로 간다는 추측에서 나온 수열

  
##### 2. 반복문으로 접근하기


>   반복적 접근법이 처음에는 "unpythonic"처럼 보일 수 있지만, 콜라츠 수열은 매우 반복적인 알고리즘이며, 이 문제를 해결하기 위한 쉬운 기능적 방법은 많지 않다. 




```python
def collatz_len(n):
  len = 0
  while True:
    len += 1
    if n==1:
      break
    elif n%2==0:
      n = int(n/2)
    else:
      n = 3*n + 1
  return len
```

```python
collatz_len(2)
```

3. **Pythonic한 방법**
  수열의 모든 요소에 함수를 적용하는 'map'을 사용하기.
  >return max(map(collatz_len, range(1, n)))

4. **옵션 2**: using al ist comprehension (아직 수업에서 다루지 않은 내용)
  > return max([collatz_len(i) for i in range(1, n)])


```python
def max_collatz_len(n):
    # return max( map(collatz_len, range(1,n)) )
    return max( [collatz_len(i) for i in range(1,n)] )
```

### 문제 3-2: 조금 더 영리하게 collatz 길이 찾기


Dictionary는 이전 결과의 캐시로 사용되며 (전달된 dictionary는 변경할 수 있으므로) 변경 사항은 호출자에게 반영됩니다.


```python
def collatz_len_fast(n, cache):
    if n==1:
      return 1
    if n in cache:
      return cache[n]
    
    if n%2==0:
      cache[n] = collatz_len_fast(n//2, cache) + 1
    else:
      cache[n] = collatz_len_fast(n*3+1, cache) + 1
    return cache[n]
```

### 1-1. 문제: 재귀적인 방법을 통해 mxn 그리드의 왼쪽 상단 모서리에서 오른쪽 하단까지의 격자 경로 수 계산하기

오른쪽 또는 아래로 움직일 수 있습니다.

m x n 그리드는 다음과 같습니다.

        (# of paths that start          (# of paths that start
         by going to the right)    +     by going down        )

참고사항: 
1. 케이스 나누기
  1. 오른쪽으로 시작하는 경로의 수
     mx(n-1) 그리드의 왼쪽 위부터 해당 그리드의 오른쪽 아래까지의 경로 수

  2. 왼쪽으로 시작하는 경로의 수
     (m-1) x n 그리드의 왼쪽 상단부터 오른쪽 하단까지의 경로의 수

2. 재귀적 접근
  재귀 문제로 해결시, 기본 케이스의 경우:
  m == 0 또는 n == 0일 때, 그리드는 실제로 1차원이므로 끝까지 가는 유일한 방법은 선을 따라 걷는 것입니다.

```python
def lattice_paths(m, n):
    if m==0 and n==0:
      return 1
    elif m==0:
      # print('01')
      return lattice_paths(m, n-1)
    elif n==0:
      # print('02')
      return lattice_paths(m-1, n)
    else:
      # print('03')
      return lattice_paths(m-1, n) + lattice_paths(m, n-1)
```

```python
lattice_paths(3,3)
```

### 3. 문제: 최대공약수 구하기

참고사항:


- tuple-packing과 unpacking 
  임시 변수를 사용하지 않고도 pair으로 구성된 업데이트 단계를 한 줄로 구할 수 있습니다.
      
      gcd(91, 35)
        a, b = 91, 35
        a, b = (35, 91 % 35) = 35, 21
        a, b = (21, 35 % 21) = 21, 14
        a, b = (14, 21 % 14) = 14,  7
        a, b = ( 7, 14 %  7) =  7,  0
        b = 0, so we return 7

```python
def gcd(a, b):
    c = a%b
    if c == 0:
      return b
    else:
      return gcd(b, c)
```
